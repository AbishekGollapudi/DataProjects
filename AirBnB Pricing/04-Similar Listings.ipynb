# Finding Similar Listings Using NearestNeighbors

```python
import pandas as pd
from sklearn.neighbors import NearestNeighbors
```

```python
df = pd.read_pickle('cleaned_df.pkl')
```

```python
df.head()
```

```python
ids = df['id']
X = df.drop(columns=['id'])
```

```python
arr = X.loc[1049].values
arr = arr.reshape(-1,1)
```

```python
X.loc[1049]
```

## NearestNeighbors Algorithm

```python
neigh = NearestNeighbors(n_neighbors=6)
neigh.fit(X.values)
```

```python
recs = neigh.kneighbors(arr.T, return_distance=False)
```

```python
df.loc[821]
```

```python
df2 = pd.read_csv('tokyo_sept.csv')
df2 = df2[['listing_url', 'id']]
```

```python
comb = df2.merge(df, on=['id'])
```

```python
comb
```

```python
recs
```

## Finding the 5 Most Similar Listings

```python
rec_df = pd.DataFrame(columns=comb.columns.tolist())
for i in recs:
    rec_df = rec_df.append(comb.iloc[i], ignore_index=True)

rec_df.iloc[1:]
```
